```python
# Import necessary libraries
import pandas as pd
from sklearn.datasets import load_digits
# Load the digits dataset from sklearn.datasets
digits = load_digits()
```

```python
# Display the target values of the dataset
digits.target
```

```python
# Display the available attributes and methods of the 'digits' dataset
dir(digits)
```

```python
# Display the unique target names of the dataset
digits.target_names
```

```python
# Create a DataFrame from the digits dataset, setting the target values as the index
df = pd.DataFrame(digits.data, digits.target)
# Display the first five rows of the DataFrame
df.head()
```

```python
# Add a 'target' column to the DataFrame with the target values
df['target'] = digits.target
# Display the first 20 rows of the DataFrame, including the new 'target' column
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3, random_state=10)
```

```python
# Import KNeighborsClassifier from sklearn.neighbors
from sklearn.neighbors import KNeighborsClassifier
# Initialize the KNeighbors Classifier with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# Display the number of samples in the training set
len(X_train)
```

```python
# Display the number of samples in the test set
len(X_test)
```

```python
# Fit the KNN model to the training data
knn.fit(X_train, y_train)
```

First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('carprices.csv')
df
```

```python
# Plotting the relationship between Mileage and Sell Price
%matplotlib inline
plt.xlabel('Mileage')
plt.ylabel('Sell Price($)')
plt.scatter(df['Mileage'], df['Sell Price($)'], color='red', marker='+')
```

```python
# Convert categorical variable 'Car Model' into dummy variables
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
# Merge the original DataFrame with the dummy variables
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# Drop the original 'Car Model' column and one of the dummy variable columns to avoid the dummy variable trap
final = merged.drop(["Car Model", "Mercedez Benz C class"], axis='columns')
final
```

```python
# Prepare the feature matrix (X) and the target vector (y)
X = final.drop('Sell Price($)', axis='columns')
y = final['Sell Price($)']
```

```python

y
```

```python
# Initialize and train the linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
# Assess the model's performance
model.score(X, y)
```

# Price of mercedez benz that is 4 yr old with mileage 45000

```python
# Making predictions with the trained model
model.predict([[45000, 4, 0, 0]])
```

# Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000, 7, 0, 1]])
```

# Using sklearn OneHotEncoder

```python
# Applying label encoding to 'Car Model'
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle['Car Model'] = le.fit_transform(dfle['Car Model'])
```

```python
# Preparing the data with label encoding applied
X = dfle[['Car Model', 'Mileage', 'Age(yrs)']].values
X
```

```python
y = dfle['Sell Price($)'].values
y
```

```python
# Applying one-hot encoding to the 'Car Model' column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([('Car Model', OneHotEncoder(), [0])], remainder='passthrough')
ct
```

```python
X = ct.fit_transform(X)
X
```

```python
# Training the model again with the newly encoded data
model.fit(X, y)
```

```python
model.predict([[45000,4,0,0,0]])
```
